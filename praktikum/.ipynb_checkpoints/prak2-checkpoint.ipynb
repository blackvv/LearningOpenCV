### Homogeneous Coordinates

Mengubah dari Cartesian Coordinates ke Homogeneous Coordinates

(x,y)→(x,y,1)

(x,y,z)→(x,y,z,1)
 
Contoh:

(2,3)→(2,3,1)

(10,−4,−2)→(10,−4,−2,1)

Mengubah dari Homogeneous Coordinates ke Cartesian Coordinates

(x,y,w)→(x/w,y/w)

(x,y,z,w)→(x/w,y/w,z/w)
 
Contoh:

(6,−2,2)→(3,−1)

(9,−24,−3,−3)→(−3,8,1)

```python
#Tampilkan cartesian coordinates dari homogeneous coordinates (6, -2, 2)
p1 = np.array([6, -2, 2])
print(p1[0:2])
#atau
print(p1[:2])
```

**Homogeneous Coordinates Properties**

Skala tidak berpengaruh, contoh  (x,y,w)=k(x,y,w) . Kedua titik  (x,y,w)  dan  (kx,ky,kw)  mewakili titik 2D Cartesian yang sama  (x/w,y/w).

**Constructing a Line**

Anggap ada 2 titik 2D Cartesian  (x1,y1)  and  (x2,y2) . Kita dapat membuat garis yang melewati keduanya dengan menggunakan cross-product pada homogeneous coordinates.

Ingat kembali persamaan garis adalah  ax+by+c=0  maka

(a,b,c)=(x1,y1,1)×(x2,y2,1)

**Intersecting Lines**

Jika ada 2 garis:  a1x+b1y+c1=0  dan  a2x+b2y+c2=0 , kita dapat mencari intersection-nya (bersilang) menggunakan cross-product berikut ini:

Langkah 1: Temukan titip intersection (persilangan) dalam homogeneous coordinates

(x,y,w)=(a1,b1,c1) × (a2,b2,c2)

Langkah 2: Ubah homogeneous coordinates menjadi Cartesian coordinates

Titik intersection dalam Cartesian coordinate  (x/w,y/w)
